```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def boxcox_transform(data):
    # Ensure the data is positive, as Box-Cox is not defined for non-positive values
    if any(x <= 0 for x in data):
        raise ValueError("All data points must be positive for Box-Cox transformation.")
    
    # Apply Box-Cox transformation
    transformed_data, best_lambda = stats.boxcox(data)
    
    # Return the transformed data and the best lambda value
    return transformed_data, best_lambda

# Example usage
data = [1, 2, 5, 10, 20, 50, 100]
transformed_data, best_lambda = boxcox_transform(data)

# Output the transformed data and lambda value
print("Transformed data:", transformed_data)
print("Optimal lambda (best lambda):", best_lambda)

# Visualizing the result (original vs transformed data)
plt.figure(figsize=(12, 6))

# Plot original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot transformed data
plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=10, edgecolor='black')
plt.title("Box-Cox Transformed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
